## Abstract

Unreliable power is a common problem among african business and households. Most of the current research on power outages has focused on the causes and  impacts of outages and not necessarily their nature/pattern. Understanding when, where and how power outages will occur is important to plan accordingly and prevent them. Our project seeks to explore hourly outage data obtained for Osun state in Nigeria at the distribution feeder level to create predicitve models for power outages in time and space based on previous records, electrical connections, and weather among others.

First, we gathered different datasets to be used for the model. This included power outage data (our target variable), weather data, transmission lines and power plants data, demographics and more. Then we analyzed and cleaned the data, we grouped and merged it until we had dataframes that could have predective power for our three questions: 1) How many outages will occur in a day? 2) How long will the duration of the outage be? and 3) Will a given hour have a power outage? We used hyperparameters to tune the bias-variance trade off, we looked at features with the most predective power and we plotted graphs to visualize the model's true outcomes vs predicted outcomes. 

Questions 1 and 2, are regression problems and we built four different models to compare their given errors: ordinary least aquares (OLS), ridge, LASSO and random forest. For question 1 the model with the best performance was Ridge, while for question 2, Random Forest outperformed the others. We defined question 3 as a classification problem and we built 3 different models: boosting, random forest and support vector machine (SVM). For the last question, random forest had the best accuracy and performance. 

Working on this project taught us the value of prediction for preventing and preparing for undesired events. We also learned how challenging it is to get and work with data, especially from developing countries. Furthermore, we found that non of our models works good enough to be reliable and advice decision makers. We still believe that if the models are further trained, with more observations and features, they could turn out to be useful and help allocate resources to local governments and electric utilities for the prevention or impact reduction of power outages. 

## Project Background
African enterprises reportedly experience power outages averaging more than 50 hours/month. According to the World Bank, unreliable power is a first-order obstacle to business in Africa. For instance, the International Monetary Fund (IMF) states that a lack of access to reliable electricity costs Nigeria an estimated $29 billion a year. Relatedly, the United Nations Sustainable Development Goal (SDG) 7 which calls for universal access to affordable, reliable, sustainable and modern energy for all by 2030 recognizes reliability as an essential component in the definition of access (SDG 7.1). 

Although power outages on the continent are prevalent, erratic and both economically and socially disruptive, the scope and pattern of these outages are not well understood both by everyday citizens and in the literature. For example, Farquharson et al. (2018) in their paper on the impact of backup diesel generator use in Sub-Saharan Africa (SSA) due to regular outages, noted that “unfortunately, without detailed data on the cause and exact timing of the outages, a more detailed analysis is not possible at this time.” This is the gap that our research project seeks to address, with a focus on the Nigerian context for which we were able to obtain data. 

Most of the current research on outages has focused on the causes and impacts of outages and not necessarily their nature/pattern. Ogunlami et al. (2021) examined the impact of electricity outages, considering primarily the performance of micro enterprise firms in selected semi-urban centres of Lagos and Ogun States, Nigeria. Their study was based on survey data collected using purposive sampling. In other words, their goal was not to model the pattern of outages but instead collect information from businessowners about their perceptions on the extent and effect of outages. As referenced above, Farquharson et al. (2018) used a Monte Carlo Analysis framework to estimate the changes in net air emissions, consumer costs and fossil energy consumption that result from the use of backup diesel generators in Sub-Saharan Africa due to regular outages. They found that the use of such backup generation led to worse outcomes than if the grid would have supplied electricity constantly. Samuel et al. (2019) noting that voltage instability and collapse are common electricity reliability issues developed a “new line stability index” (NLSI_1) for predicting voltage collapse on the Nigerian grid. Their index accounted for switching logic based on difference in voltage angle between two load buses. Maliki et al. (2011) developed regression and artificial neural network models for predicting electrical power generation in Nigeria.

The above papers are helpful contributions to the field. However, they do not directly address the outages themselves and offer insights that can aid planning in the presence of these outages. Quantitative analyses geared towards predicting the frequency of outages, the duration of outages, and occurrence of outages among others can help address the perceived erratic nature of the outages which is often most frustrating. It will be highly beneficial to know, for example, if, when and for how long one might lose power.

## Input Data Description

Nine datasets were used in this project: 

1. **Outage logs at the distribution feeder level**: Records of all outages that occurred in 2020 in Osun state were obtained directly from engineers in the distribution company in excel format. Workers at the distribution company record the outages daily and compile them into monthly logs.

2. **Weather Data**: This was obtained from the Trans-African Hydro-Meteorological Observatory (TAHMO) project website: https://tahmo.org/. The TAHMO team seeks to develop a dense network of hydro-meteorological monitoring stations in sub-Saharan Africa – one every 30 km which entails the installation of 20,000 stations across the continent. Currently, they have sensors in several locations across Africa and make available the data from their sensors upon request. 

3. **Distribution and Transmission Data**: Two different datasets containing shapefiles of the existing and planned transmission and distribution lines in Nigeria were obtained from the following links: (https://datacatalog.worldbank.org/search/dataset/0042323) and (https://energydata.info/dataset/nigeria-electricity-transmission-network-2015/resource/2ac1f255-8c71-45f9-a7e7-45c317e454bb). Both datasets were credited to the World Bank. The World Bank compiled the data as part of an Africa Infrastructure Country Diagnostic (AICD) study, and collected them by drawing on multiple sources, including regional power  documents and maps from World Bank project documents.

4. Distribution and Transmission Data (as explained above)

5. **Power plants Data**: A dataset containing information (such as geometric location) of the existing power plants in Nigeria was downloaded in a zip file from the following link: https://datacatalog.worldbank.org/search/dataset/0042336. This data was also collected by The World Bank.

6. **Daily Energy Operational reports**: The sixth dataset is a daily operational report from the Nigeria Electricity System Operator (NESO) which provides interesting information such as daily total energy generated and daily peak generation capacity. The NESO website (https://www.nsong.org/Library) uploads this information daily as pdf documents. Unfortunately, we could not get a collated .csv document from the company and had to manually extract the information ourselves. Information from a sample pdf has been extracted for the description below. It is important to note that although values such as daily total energy generated and highest frequency on the grid will not be available prior to the day in question, forecasted generation values or assumptions (as is the case with weather data) can be used for predictions instead. In addition,one can imagine a use-case for our model where outages are retroactively predicted before manual records come in or if they are not collated. Also, different predictions of frequency and voltage could be used to predict a likely range of outages. 

7. **Nigeria State Boundaries Data**: The shapefile for nigeria's state boundaries was obtained from the GRID3 (Geo-Referenced Infrastructure and Demographic Data for Development) website at this link: https://data.grid3.org/datasets/GRID3::grid3-nigeria-state-boundaries/about. According to their website, "GRID3  works with countries to generate, validate and use geospatial data on population, settlements, infrastructure, and boundaries." They note that, "This work *[to produce the data]* has been undertaken as part of the Geo-referenced Infrastructure and Demographic Data for Development (GRID3) initiative in NGA. Funding for the development and dissemination of this data set was provided by the Bill & Melinda Gates Foundation and UK aid from the UK government. The following contributed to the production of the this data set: eHealth Africa, Proxy Logics, GeoBoundaries, and CIESIN, Columbia University" This dataset was mainly used to obtain the boundaries of Osun State.

8. **Nigeria Local Government Area (LGA) Boundaries Data**: The shapefile for Nigeria's LGA boundaries was also obtained from the GRID3 website here: https://data.grid3.org/datasets/GRID3::grid3-nigeria-local-government-area-boundaries/about

9. **Socioeconomic vulnerability score per LGA**: This was downloaded as a csv from the GRID3 website as well: https://data.grid3.org/datasets/GRID3::grid3-nigeria-socioeconomic-vulnerability-score-per-lga/about. The socioeconomic vulnerability score ranged from 1 to 5, with 5 indicating the highest level of vulnerability. The score was created via a principal component analysis (PCA) with a list of indicators of socio-economic status at the household level (non-farm employment, hectares of farmland owned, number of livestock owned, educational attainment), housing type (material for roof, floor, and walls), food security (proxied by child wasting), financial inclusion (household member has a bank account), and domestic violence (household experienced physical, sexual, or emotional abuse of women at last once in a year). This index also considered urban and rural differences by conducting the PCA for urban and rural settings and combining results for a national index. Different sources such as the World Health Organization, the Independent National Electoral Commission of Nigeria (INEC), the eHealth Africa path to polio eradication in Nigeria project and GRID3 were cited for the records in the .csv.

## Data Cleaning 

Due to the multiplicity of datasets we used in this project, significant data cleaning was required. Several of these datasets were neither created for the purposes we had nor designed to be combined with each other. The notebooks detailing the data cleaning process for the individual datasets or in some cases, groups of datasets have been uploaded in our project zipfile. Here, we present a detailed summmary and include images of relevant plots. 

- **Outages Data Cleaning**: 

As mentioned earlier, we received the 2020 Osun region distribution feeder outage data as an excel file, with separate sheets for each month and feeder voltage. We imported this file and combined the sheets into a single outage data frame. Due to the fact that the records had been manually entered, they required substantial data cleaning. 

In order to do this, we identified unique values for each column--including null and nonsensical values--and dropped observations for which we could not determine the appropriate value, and cleaned those which were simply entered differently from others. For example, the date information required extensive cleaning as some of the records had been entered in the format of YYYY-MM-DD, while others had been entered as YYYY-DD-MM. To address this we wrote an algorithm that determined which value was the day and which was the month, and created a new column with the corrected date. Additionally, we removed a small number of records which were from 2019.

Feeder names also required considerable cleaning, as a given feeder’s name often varied substantially across observations. For example, the following formats all seemed to represent the same feeder; 'Ipetu-ijesa 33KV', 'Ipetu-Jesa 33KV', 'Ipetu-Ijesa 33KV', 'Ipetu Ijesa 33KV'. The main issues had to do with the inclusion of multiple spaces or special symbols like dashes, capitalization, and in some cases, the spellings. The data lacked unique IDs for each feeder, and we lacked access to the true cleaned feeder names for a significant portion of the project. We undertook this cleaning based on our knowledge of and research into locations in the Osun region after which the feeders are named. Eventually, we obtained a list of feeder names from the distribution company and adjusted the remaining discrepancies.

In addition to these columns, the business hub, nature/cause of outage, type of outage, relay target, and load loss columns required cleaning. 

Also, we initially considered using the number of customers served by a given feeder as a feature. However, in investigating the data, we found that feeders largely lacked a unique value for total number of customers served, and instead the value typically varied across observations, with a low portion of repeated values for total customers served for a given feeder, as shown in the table below. Due to the uncertainty around the reason for the variation in total customers served for a given feeder, we decided to exclude it from the models. 

- **Weather Data Cleaning**: 

As discussed earlier, a significant challenge we had with the weather data was finding a station close enough to our study location with sufficient data for our modeling purposes. Once we found a suitable proxy, cleaning operations were carried out. Primary of which was the conversion of the timestamp in the weather data from UTC timezone to West Africa/Lagos timezone (the timezone the power outage data was in, 'Nigerian Time'). Africa / Lagos time zone was selected over UTC for easy interpretation of time of day predictions. In other words, we wanted our results to be easily interpretable in the timezone of the study location. The weather data was then grouped by the hour with weather characteristics at the top of the hour chosen as we did not notice significant variation within hours. **Caveats and/or reasons results may be flawed**

The Question 1 model which predicts the number of outages per day has a very low R squared value, indicating very low predictive power. As shown in the plot of oberved vs. predicted values, the model generally performs best for days with average outage values (70-90), underpredicts the nubmer of outages on days that in fact experienced a large number of outages, and overpredicts the number of outages on days that in fact experienced a low number of outages. This may be in part because it was trained on relatively few data points, as we only had outage observation data for 2020, meaning that we only had 366 data points to include in the combined train and test dataset. It is possible that the model could be improved if we were to incorporate data for additional years, and/or incorporate additional features. 

For Question 2 which predicts outage duration, though the best performing model exhibiting a better R squared value relative to Question 1, it still had relatively low predictive power. As seen in the predicted vs. observed y-value plot, the model tends to underestimate the duration of longer-lasting outages, and overestimate the duration of shorter outages. As mentioned above,  future work may improve on this model by aggregating records for outages that span multiple days, rather than using the duration of a given outage on a given day; because the outage records lacked unique IDs and there were significant challenges involved in aggregating records, we did not undertake this effort. 

The Question 3 model which predicts whether or not an outage occurs for a given hour was also limited, in that it had a relatively low accuracy score. Additionally, it was biased toward false positives, as it had a low precision score (i.e., produced many false positives) but a high recall score (i.e., produced very few false negatives). This indicates that the model errs toward mistakenly predicting that an outage will occur when it actually won't, rather than mistakenly predicting that an outage won't occur. This could be valuable if the utility and customers would rather know if it's possible that an outage may occur at a given time, and don't want to be caught offguard by an unexpected outage (i.e., a false negative).However, if the utility would rather be more conservative and avoid notifying customers that outages may occur only to have the outage never manifest, this model likely wouldn't be very useful.
